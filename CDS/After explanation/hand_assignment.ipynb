```python
visited_set = set()
curve_points = set()
```

```python
# eight nearest neighbour
def eight_nearest_neighbour(i,j):
  
  if i-1 >= 0 and j-1 >= 0 and imgGray[i-1][j-1] == 1:
    visited_set.add((i-1,j-1)) 
    return True,i-1,j-1
  elif i-1 >= 0 and j >= 0 and imgGray[i-1][j] == 1:
    visited_set.add((i-1,j))
    return True,i-1,j
  elif i-1 >= 0 and j+1 >= 0 and imgGray[i-1][j+1] == 1:
    visited_set.add((i-1,j+1))
    return True,i-1,j+1

  elif i >= 0 and j-1 >= 0 and imgGray[i][j-1] == 1:
    visited_set.add((i,j-1))
    return True,i,j-1
  elif i >= 0 and j+1 != 0 and imgGray[i][j+1] == 1:
    visited_set.add((i,j+1))
    return True,i,j

  elif i+1 != 0 and j-1 != 0 and imgGray[i+1][j-1] == 1:
    visited_set.add((i+1,j-1))
    return True,i+1,j-1
  elif i+1 != 0 and j != 0 and imgGray[i+1][j] == 1:
    visited_set.add((i+1,j))
    return True,i+1,j
  elif i+1 != 0 and j+1 != 0 and imgGray[i+1][j+1] == 1:
    visited_set.add((i+1,j+1))
    return True,i+1,j+1 
  else:
    return False,i,j
```

```python
# Iterate over all the pixels
for i in range(width):
      for j in range(height):
        if (i,j) in visited_set:
            continue
        visited_set.add((i,j))
        if (imgGray[i][j] == 1):
            curve_points.append((i,j))
            found_neighbour = True
            ni = i
            nj = j
            while(found_neighbour == True):        
                found_neighbour,ni,nj = eight_nearest_neighbour(ni,nj)
                if (found_neighbour == True):
                    curve_points.append((ni,nj))
```
